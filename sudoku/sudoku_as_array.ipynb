# Sodoku as Array

```python
import sys
sys.path.append("..")
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from helpers.data_prep_and_print import print_df
from sudoku.data_conversions import QUIZ_ATTR, SOLUTION_ATTR
```

## Create Spark Session

```python
spark = (SparkSession
       .builder
       .appName("Sudoku as Array")
       .getOrCreate())
spark.sparkContext.setLogLevel("ERROR")
```

## Create a DataFrame using an ifered Schema 

```python
path = "../sudoku_data/"
df = spark.read.option("header", "true") \
       .option("inferSchema", "true") \
       .csv(path+"sudoku_1_mio_sequenced_columns_with_difficulty_and_num_solutions.csv")
df = df.limit(10)
#print_df (df)
```

## Build the Vector / Array of all Features and labels

```python

quiz_assembler = VectorAssembler(inputCols=QUIZ_ATTR, outputCol="quizzes")
solution_assembler = VectorAssembler(inputCols=SOLUTION_ATTR, outputCol="solutions")
df_result = solution_assembler.transform(quiz_assembler.transform(df))
df_result = df_result.select("quizzes", "solutions")
print_df (df_result)
```
